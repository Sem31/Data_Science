```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.arange(6)
y = x*2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r');
```

```python
#with --
plt.plot(x,y,'r--');
```

```python
#with *-
plt.plot(x,y,'b*-');
```

```python
#labels
plt.plot(x,y,'b*-');
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Titel');
```

```python
plt.subplot(1,2,1) #subplot(nrows, ncols, plot no.)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b*-');
```

```python
fig = plt.figure(figsize=(8,4))
axe = fig.add_axes([0,0,1,1])
axe.plot(x,y,'b');
axe.set_xlabel('X Label')
axe.set_ylabel('Y Label')
axe.set_title('Title');
```

```python
fig = plt.figure(figsize=(8,4))
axe = fig.add_axes([0,0,1,1])
axe.plot(x,y,'b');
axe.set_xlabel('outer Label')
axe.set_ylabel('outer Label')
axe.set_title('outer Title');
axe1 = fig.add_axes([0.1,0.6,0.3,0.3])
axe1.plot(y,x,'r*-');
axe1.set_xlabel('inner Label')
axe1.set_ylabel('inner Label')
axe1.set_title('inner Title');
```

```python
fig,axes = plt.subplots(1,2)
for ax in axes:
    ax.plot(x,y,'r--')
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_title('Title')
plt.tight_layout()
```

```python
fig.savefig('k.jpg',dpi = 200)
```
